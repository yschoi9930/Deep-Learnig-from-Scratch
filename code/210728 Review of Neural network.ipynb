```python
class Sigmoid :
    def __init__(self):
        self.params = []
        
    def forward(self, x) :
        return 1 / (1+np.exp(-x))
```

```python
class Affine :
    def __init__(self, W, b) :
        self.params = [W,b]
    
    def forward(self, x) :
        W, b = self.params
        out = np.matmul(x,W) +b
        return out
```

```python
class TwoLayerNet:
    def __init__(self, input_size, hidden_size, output_size) :
        I, H, O = input_size, hidden_size, output_size
        
        # 가중치와 편향 초기화
        W1 = np.random.randn(I,H)
        b1 = np.random.randn(H)
        W2 = np.random.randn(H,O)
        b2 = np.random.randn(O)
        
        # 계층 생성
        self.layers = [
            Affine(W1, b1),
            Sigmoid(),
            Affine(W2, b2)
        ]
        # 각 함수(레이어)별로 매개변수 지정 -> Affine.params = [W1,b2] 이런 식으로 저장됨
        
        # 모든 가중치를 리스트에 모은다
        self.params = []
        for layer in self.layers :
            self.params += layer.params
        
        # TwoLayerNet.params에 Affine.params, Sigmoid.params 등이 저장됨
        
    def predict(self, x) :
        for layer in self.layers:
            x = layer.forward(x)
        return x
        
        # 입력데이터 x가 Affine.forward, Sigmoid.forward 매소드로 전달됨
```

```python
class Matmul :
    def __init__(self, W) :
        self.params = [W]
        self.grads = [np.zeros_like(W)]
        self.x = None
    
    def forward(self, x) :
        W, = self.params
        out = np.matmul(x, W)
        self.x = x
        return out
    
    def backward(self, dout) :
        W, = self.params
        dx = np.matmul(dout, W.T)
        dW = np.matmul(self.x.T, dout)
        self.grads[0][...] = dW
        return dx
```

```python
class TwoLAyerNet :
    def __init__(self,input_size, hidden_size, output_size) :
        I,H,O = input_size, hidden_size, output_size
        
        # 가중치와 편향 초기화
        W1 = 0.01*np.random.randn(I,H)
        b1 = np.zeros(H)
        W2 = 0.01 * np.random.randn(H,O)
        b2 = np.zeros(O)
        
        # 계층생성
        self.layers = [
            Affine(W1,b1),
            Sigmoid(),
            Affine(W2,b2)
        ]
        self.loss_layer = SoftmaxWithLoss()


        # 모든 가중치와 기울기를 리스트에 모은다
        self.params, self.grads = [], []
        for layer in self.layers :
            self.params += layer.params
            self.grads +=layer.grads
            
    def predict(self, x):
        for layer in self.layers:
            x = layer.forward(x)
        return x
        
    def forward(self, x, t):
        score = self.predict(x)
        loss = self.loss_layer.forward(score, t)
        return loss


    def backward(self, dout=1):
        dout = self.loss_layer.backward(dout)
        for layer in reversed(self.layers) :
            dout = layer.backward(dout)
        return dout
```

```python
# 하이퍼 파라미터 설정
max_epoch = 300
batch_size = 30
hidden_size = 10
learning_rate = 1.0

# 데이터 읽기, 모델과 옵티마이저 생성
x, t = spiral.load_data()
model = TwoLAyerNet(input_size=2, hidden_size=hidden_size, output_size=3)
optimizer = SGD(lr = learning_rate)


# 학습에 사용하는 변수
data_size = len(x)
max_iters = data_size // batch_size # 몫 구하기
total_loss = 0
loss_count = 0
loss_list = []

for epoch in range(max_epoch) :
    # 데이터 두섞기
    idx = np.random.permutation(data_size)
    x = x[idx]
    t = t[idx]
    
    for iters in range(max_iters) :
        batch_x = x[iters*batch_size:(iters+1)*batch_size] # x[0:10]...
        batch_t = t[iters*batch_size:(iters+1)*batch_size]

      # 기울기를 구해 매개변수 갱신
        loss = model.forward(batch_x, batch_t)
        model.backward()
        optimizer.update(model.params, model.grads)
        
        total_loss += loss
        loss_count += 1

        # 정기적으로 학습경과 출력
        if (iters+1) % 10 ==0 :
            avg_loss = total_loss / loss_count
            print('| 에폭 %d | 반복 %d / %d | 손실 %.2f' % (epoch+1, iters+1, max_iters, avg_loss))
            loss_list.append(avg_loss)
            total_loss, loss_count = 0, 0
```
